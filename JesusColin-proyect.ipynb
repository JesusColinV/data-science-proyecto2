```python
import csv
path = "synergy_logistics_database.csv"
```

```python
with open(path,'r') as data:
    lector = csv.DictReader(data)
    mydata = [linea for linea in lector]
```

Para cada uno de las variables obtenemos el numero: es decir la cantidad de viajes que se realizaron, el value: Es decir, el valor de la carga y se obtiene un promedio para considerar aquellos que son más frecuentes pero tambien los de mayor valor.

```python
len(mydata)
category = {
'road':{},
'origin':{},
'destination':{},
'company_name':{},
'date':{},
'transport_mode':{},
'product':{},
}

for ruta in mydata:
    try:
        category['road'][ruta['origin']+'-'+ruta['destination']]={
            'number':category['road'][ruta['origin']+'-'+ruta['destination']]['number']+1,
            'value':category['road'][ruta['origin']+'-'+ruta['destination']]['value']+int(ruta['total_value']),
            'average':(category['road'][ruta['origin']+'-'+ruta['destination']]['number']+1)/(category['road'][ruta['origin']+'-'+ruta['destination']]['value']+int(ruta['total_value']))}
    except:
        category['road'][ruta['origin']+'-'+ruta['destination']]={
            'number':1,
            'value':int(ruta['total_value']),
            'average':int(ruta['total_value'])/1}

    for categoria in list(category.keys())[1:]:
        try:
            category[categoria][ruta[categoria]]={
                'number':category[categoria][ruta[categoria]]['number']+1,
                'value':category[categoria][ruta[categoria]]['value']+int(ruta['total_value']),
                'average':(category[categoria][ruta[categoria]]['number']+1)/(category[categoria][ruta[categoria]]['value']+int(ruta['total_value']))}
        except:
            category[categoria][ruta[categoria]]={
                'number':1,
                'value':int(ruta['total_value']),
                'average':int(ruta['total_value'])/1}
```

```python
def order(diccionario,orden):
    arreglo=[[k,v] for k,v in diccionario.items()]
    # Calculamos la longitud del arreglo para tener un código más limpio
    longitud = len(arreglo)
    # Recorremos todo el arreglo
    for i in range(longitud):
        # Dentro del ciclo, volvemos a recorrerlo. Pero ahora hasta el penúltimo elemento
        for indice_actual in range(longitud - 1):
            indice_siguiente_elemento = indice_actual + 1
            # Si el actual es mayor que el siguiente, ...
            # Nota: para un orden inverso, cambia `>` por `<`
            if orden == 'min':
              if arreglo[indice_actual][1] > arreglo[indice_siguiente_elemento][1]:
                  # ... intercambiamos los elementos
                  arreglo[indice_siguiente_elemento], arreglo[indice_actual] = arreglo[indice_actual], arreglo[indice_siguiente_elemento]
            elif orden == 'max':
              if arreglo[indice_actual][1] < arreglo[indice_siguiente_elemento][1]:
                  # ... intercambiamos los elementos
                  arreglo[indice_siguiente_elemento], arreglo[indice_actual] = arreglo[indice_actual], arreglo[indice_siguiente_elemento]
    #print(arreglo)
    return arreglo[0:5]
```

```python
for categoria in category.keys():
    print(f"\nEstos son los mejores en la categoria de {categoria}")
    for i in order({ k:v['average'] for k,v in category[categoria].items()},'max'):
        print(i)
```
